input data

```python
X=df.iloc[:,:-1]
X
```

output data

```python
y=df.iloc[:,-1]
```

Preprocessing

```python

scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
y.value_counts()
```

Model selection

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

Different model building

```python
# acc=[]
from sklearn.metrics import accuracy_score
knn=KNeighborsClassifier()
sv=SVC()
nb=GaussianNB()
tr=DecisionTreeClassifier()
rf=RandomForestClassifier()
ada=AdaBoostClassifier()
xg=XGBClassifier(n_estimators=500,learning_rate=0.05,max_depth=6)
models=[knn,sv,nb,tr,rf,ada,xg]
for model in models:
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  print("*************************",model,"***********************")
  print("***************classification_report*********************",classification_report(y_test,y_pred))
  print("***************ConfusionMatrixDisplay*********************",ConfusionMatrixDisplay.from_predictions(y_test,y_pred))


```

Feature dropping

```python
df.corr()
```

```python
df.drop(['height_cm','age'],axis=1)
```

```python
X1=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
scaler=MinMaxScaler()
X_new=scaler.fit_transform(X1)
```

```python
X_trainn,X_testn,y_trainn,y_tsetn=train_test_split(X_new,y,test_size=0.15,random_state=1)
```

```python
acc=[]

for model in models:
  model.fit(X_trainn,y_trainn)
  y_pred1=model.predict(X_testn)
  print("*************************",model,"***********************")
  print("***************classification_report*********************",classification_report(y_tsetn,y_pred1))
  acc.append(accuracy_score(y_tsetn,y_pred1))
```

Model performance graph

```python
modelnm=['knn','sv','nb','tr','rf','ada','xg']
plt.bar(modelnm,acc,)
```

XGBClassifier show best accuracy
